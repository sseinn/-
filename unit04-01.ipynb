```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10, 20, 30, 40]) # 리스트에 입력한 값은 y축 값이다
```

```python
plt.plot([1, 2, 3, 4], [12, 43, 25, 15]) # 리스트에 입력한 첫 번째 값 : x축, 두 번째 값 : y축
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('plotting') # 그래프에 제목 넣기
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
# 그래프에 범례 넣기

import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10, 20, 30, 40], label = 'asc')
plt.plot([40, 30, 20, 10], label = 'dec')
plt.legend()
plt.show()
```

```python
# 그래프 색상 바꾸기

import matplotlib.pyplot as plt
plt.title('color')
plt.plot([10, 20, 30, 40], color = 'skyblue', label = 'skyblue')
plt.plot([40, 30, 20, 10], 'pink', label = 'pink')
plt.legend()
plt.show()
```

```python
# 그래프 선 모양 바꾸기

import matplotlib.pyplot as plt
plt.title('linestyle')
plt.plot([10, 20, 30, 40], color = 'r', linestyle = '--', label = 'dashed')
plt.plot([40, 30, 20, 10], color = 'g', ls = ':', label = 'dotted')
plt.legend()
plt.show()
```

```python
# 마커 모양 바꾸기

import matplotlib.pyplot as plt
plt.title('marker')
plt.plot([10, 20, 30, 40], 'r.', label = 'circle')
plt.plot([40, 30, 20, 10], 'g^', label = 'triangle up')
plt.legend()
plt.show()
```
