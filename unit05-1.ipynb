```python
import random

dice = []

for i in range(5):
    dice.append(random.randint(1, 6))
print(dice)
```

```python
import matplotlib.pyplot as plt
plt.hist(dice, bins = 6) # bins : 가로축의 구간 개수 설정 속성
plt.show()
```

```python
# 기온 데이터를 히스토그램으로 표현하기 -> 빨간색, 구간은 100

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

plt.hist(result, bins = 100, color = 'r')
plt.show()
```
