```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = [] # 최고 기온을 저장할 리스트 생성

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
print(result)
```

```python
# 8월의 최고 기온 데이터 

import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            result.append(float(row[-1]))

plt.plot(result, 'hotpink')
plt.show()
```

```python
# 매년 2월 14일의 최고 기온 데이터

import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
            result.append(float(row[-1]))

plt.plot(result, 'hotpink')
plt.show()
```

```python
# 1983년 이후 매년 2월 14일의 최고기온, 최저기온  추출해서 그래프 그려보기

import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        if float(row[0].split('-')[0]) >= 1983:
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.plot(high, 'hotpink', label = 'high')
plt.plot(low, 'skyblue', label = 'low')
plt.rc('font', family = 'Malgun Gothic')
plt.title('내 생일의 기온 변화 그래프')

plt.rcParams['axes.unicode_minus'] = False # y축 마이너스 기호가 깨질 때
plt.legend()

plt.show()
```
